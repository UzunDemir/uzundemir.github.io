```python
import numpy as np

# Создание матрицы из двух столбцов
column1 = np.array([1, 1, 1, 1, 1])
column2 = np.array([1, 2, 3, 4, 5])
matrix = np.column_stack((column1, column2))
y = np.array([45, 40, 35, 30, 25])
print(matrix, '\n')
print(y)
```

```python
# Транспонирование матрицы
transposed_matrix = np.transpose(matrix)
print(transposed_matrix)
```

```python
# Перемножение матриц
result = np.dot(transposed_matrix, matrix)
print(result)
```

```python
# Нахождение обратной матрицы
try:
    inverse_matrix = np.linalg.inv(result)
    print(inverse_matrix)
except np.linalg.LinAlgError:
    print("Обратная матрица не существует.")
```

```python
import numpy as np

# Определение матрицы X
X = np.array([[1, 1, 2],
              [1, 2, 4],
              [1, 3, 6],
              [1, 4, 8],
              [1, 5, 10]])
print('matrix X', '\n', X, '\n')

# Опрcvеделение вектора Y
Y = np.array([6, 11, 16, 21, 26])
print('vector Y', '\n', Y, '\n')

# Вычисление транспонированной матрицы XT
XT = X.T
print('transposed matrix XT', '\n', XT, '\n')

# Вычисление матрицы XTX
XTX = np.dot(XT, X)
print('matrix XTX', '\n', XTX, '\n')

determinant = np.linalg.det(XTX)
print("Определитель матрицы:", determinant, '\n')

# Задание значения α
alpha = 0.1

# Создание единичной матрицы размером k x k
I = np.eye(XTX.shape[1])
print("Создание единичной матрицы:", I, '\n')

# Дополнение матрицы XTX+αI
XTX_alphaI = XTX + alpha * I
print('Дополнение матрицы XTX+αI', '\n', XTX_alphaI, '\n')

determinant = np.linalg.det(XTX_alphaI)
print("Определитель матрицы:", determinant, '\n') 

# Вычисление обратной матрицы (XTX+αI)-1
XTX_alphaI_inv = np.linalg.inv(XTX_alphaI)
print('Вычисление обратной матрицы (XTX+αI)-1', '\n', XTX_alphaI_inv, '\n')



# Вычисление вектора XTy
XTy = np.dot(XT, Y)
print('Вычисление вектора XTy', '\n', XTy, '\n')

# Вычисление вектора весов w
w = np.dot(XTX_alphaI_inv, XTy)
print('Вычисление вектора весов w', '\n', w, '\n')

# Вывод уравнения регрессии
print("Уравнение регрессии: Y =", '\n', w[0],'+', w[1], "* X1 +", w[2], "* X2", '\n')

print('----------------Метрики-----------------', '\n')
###################################### MSE
# Предсказанные значения
y_pred = np.dot(X, w)

# Расчет среднеквадратичной ошибки
mse = np.mean((y_pred - Y) ** 2)
print('MSE = ', mse)

##################################### R-squared
# Расчет суммы квадратов отклонений
ss_total = np.sum((Y - np.mean(Y)) ** 2)

# Расчет суммы квадратов остатков
ss_residual = np.sum((Y - y_pred) ** 2)

# Расчет коэффициента детерминации
r_squared = 1 - (ss_residual / ss_total)
print('R-squared = ', r_squared)

##################################### Mean Absolute Error, MAE
# Предсказанные значения
y_pred = np.dot(X, w)

# Расчет средней абсолютной ошибки
mae = np.mean(np.abs(y_pred - Y))
print('Mean Absolute Error = ', mae)

#################################### Adjusted R-squared
# Предсказанные значения
y_pred = np.dot(X, w)

# Расчет суммы квадратов остатков
ss_residual = np.sum((Y - y_pred) ** 2)

# Расчет суммы квадратов отклонений
ss_total = np.sum((Y - np.mean(Y)) ** 2)

# Расчет числа независимых переменных
n = X.shape[0]
p = X.shape[1]

# Расчет скорректированного коэффициента детерминации
adjusted_r_squared = 1 - (ss_residual / (n - p - 1)) / (ss_total / (n - 1))
print('Adjusted R-squared = ', adjusted_r_squared)
print('\n')
print('----------------------------------------', '\n')
```

Коэффициент детерминации (R-squared) является метрикой, которая измеряет долю объясненной вариации в данных моделью линейной регрессии. 
Значение R-squared находится в диапазоне от 0 до 1, где:
* Значение 0 означает, что модель не объясняет вариацию в данных.
* Значение 1 означает, что модель полностью объясняет вариацию в данных.

Таким образом, чем ближе значение R-squared к 1, тем лучше модель линейной регрессии соответствует данным и лучше объясняет вариацию в целевой переменной.

Adjusted R-squared (скорректированный коэффициент детерминации) является модификацией обычного R-squared, которая учитывает количество предикторов (факторов) в модели. Он позволяет сравнивать модели с разным количеством предикторов и учитывает степень свободы модели.
Adjusted R-squared увеличивается, когда добавляются предикторы, которые значимо улучшают объяснение вариации в данных. Он также штрафует за добавление незначимых предикторов, которые не улучшают модель.
Таким образом, когда значение Adjusted R-squared близко к 1, это указывает на то, что модель хорошо соответствует данным и объясняет вариацию в целевой переменной, учитывая количество предикторов. Более высокое значение Adjusted R-squared обычно указывает на более предсказательную модель.
