```python
import sympy as sp
from sympy.plotting import plot
sp.init_printing(use_unicode=False, wrap_line=False, no_global=True)

import matplotlib.pyplot as plt
import numpy as np
     
```

```python
x = sp.Symbol('x')
```

```python
f = 3*x + 5
f
```

```python
points_X = [2, 4, 8] #
points_Y = [12.258, 17.24, 30.151] #

X = np.arange(0, 10, 0.01) #
Y = [3*x + 5 for x in X]

plt.scatter(points_X, points_Y, c='r')
plt.plot(X, Y, 'b')
plt.show()
```

image.png

```python
i_y = [] # y на графике нашей функции для нахождения значения среднеквадратичной
# ошибки для данных наборов точек
counter_per_step = 0
for i in points_X:
  print(i)
  i_y.append(f.subs(x, i))
print(i_y)

# MSE
def mse():
  global counter_per_step
  for i in range(len(i_y)):
    #print(i)
    counter_per_step += (i_y[i] - points_Y[i])**2
    print(i_y[i], '-', points_Y[i], '+', '=', counter_per_step)
  MSE = counter_per_step / len(i_y)
  return MSE
                               

mse()
```
